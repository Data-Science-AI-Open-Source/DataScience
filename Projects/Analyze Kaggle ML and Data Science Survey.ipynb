```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
response=pd.read_csv('../input/multipleChoiceResponses.csv',encoding='ISO-8859-1')
response.head()
```

 # Gender Split

# Compensation

```python
salary = response[["CompensationAmount", "CompensationCurrency"]]
salary = salary.dropna(how='any')
salary["CompensationAmount"] = pd.to_numeric(salary['CompensationAmount'].str.replace(',','').str.replace('-',''))
plt.subplots(figsize=(15,8))
salary=salary[salary['CompensationCurrency']== "USD"]
salary = salary.dropna(how='any')
salary = salary[salary['CompensationAmount'] > 1000]
salary = salary[salary['CompensationAmount'] < 300000]
sns.distplot(salary['CompensationAmount'])
plt.title('Salary Distribution',size=15)
plt.show()
```
